# Análisis de Ventas & Productos Destacados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv(r"C:\Users\cpame\OneDrive\Escritorio\Proyectos_Demo\supermarketsales\data\raw\SuperMarketAnalysis.csv")

# Display the first few rows of the dataset
print('The first 5 rows of the dataset:')
display(data.head())

print('\n ----------------------------------------------------')

# Display summary statistics
print('Summary statistics of the dataset:')
display(data.describe())

print('\n ----------------------------------------------------')    
# Display information about the dataset
print('Information about the dataset:')
display(data.info())

print('\n ----------------------------------------------------')
# Vier the size of the dataset
print('The size of the dataset:')
display(data.shape) 

print('\n ----------------------------------------------------')
# Check for missing values
print('Missing values in the dataset:')
display(data.isnull().sum())

print('\n ----------------------------------------------------')
# Visualize a sample of the data
print('A sample of the data:')
display(data.sample(10))



```

```python
# Eliminamos espacios y reemplazamos por guiones bajos
data.columns = data.columns.str.strip().str.replace(' ', '_')

# Cambiamos las mayusculas a minúsculas
data.columns = data.columns.str.lower()
print(data.columns)
```

```python
# To be sure the sample size is appropriate, we can check the number of transactions. 
print('Total of unique invoices:', data['invoice_id'].nunique())
print('Minimum or earlier invoice date:', data['date'].min())
print('Maximum or later invoice date:', data['date'].max())

# We will also check the total number of sales per day to see if there are any trends or patterns in the data.
total_sales_per_day = data.groupby('date')['invoice_id'].nunique().reset_index()
avg_sales_per_day = total_sales_per_day['invoice_id'].mean()

print('Average number of sales per day:', avg_sales_per_day)
```

### Observations:

## 1. What are we selling?

```python
# Calculamos el margen o ganancia que deja cada linea de producto.

product_profits = (data.groupby('product_line')['gross_income'].sum().sort_values(ascending=False).reset_index()) 


# Visualizamos los resultados
plt.figure(figsize=(10,6))

sns.barplot(data = product_profits, y='product_line', x='gross_income', hue='product_line', palette="viridis")
plt.title('Total Ganancia por Linea de Producto')
plt.xlabel('Ganancia Total')
plt.ylabel('Linea de Producto')
plt.show()
```

```python
# Let's analyze the numbers of each product line gross
display(product_profits)
```

### Which products generate more money?

```python
# Let's Calculate net sales without taxes.
data['net_sales'] = data['unit_price'] * data['quantity']

# We create a new table with the net sales and gross income by product line
sales_summary = data.groupby('product_line')[['net_sales','gross_income']].sum()
sales_summary['margin'] = sales_summary['gross_income'] / sales_summary['net_sales']
sales_summary = sales_summary.sort_values('gross_income', ascending=False)
sales_summary = sales_summary.reset_index()
display(sales_summary)

```

## Insights:
### We can see here based on the numbers that the margin remains the same. Gross income = 5% of the sales; so we cannot answer which products leave less margin and which products sell less but leave a bigger margin.

### I tried to analyze margin by product, but I discovered that the dataset uses a fixed markup of 5%, so margin is not an explanatory variable. So I will now focus the analysis on sales volume.

### Now we will find out more information but related to the absolute "Sales"value. What has an impact on sales...

```python
# Now we can calculate the percentage of sales for each product line in base to the total sales.
sales_summary['sales_pct'] = sales_summary['net_sales'] / sales_summary['net_sales'].sum() * 100
display (sales_summary)
```

### Insights:
### As suspected, sales are well balanced and distributed. We will now proceed to chec the cummulative percentage, to see in how many categories we have 80% of sales.

```python
# We now calculate the cumulative percentage of sales for each product line.
sales_summary['cumulative_sales_pct'] = sales_summary['sales_pct'].cumsum()
display(sales_summary)
```

### Insights:
### Here we can see that we need 5 categories to reach 80+%. This means that there is no dominant category, we need all of them.
### Pros: Diversificated risk
### Cons: No clear "cash cow" 

### With this part of the analysis we now know: WHAT ARE WE SELLING.


## 2. How are we selling ?

#### Price Vs Volume?

```python
# What moves the sales price or volume of sales? We can analyze the relationship between net sales and gross income for each product line.
# We need avg price, units sold and total sales for each product line.
price_volume_analysis = data.groupby('product_line').agg(avg_price=('unit_price', 'mean'), units_sold=('quantity','sum'), net_sales=('net_sales', 'sum')).reset_index()
price_volume_analysis = price_volume_analysis.sort_values('net_sales', ascending=False)
print('Price and Volume Analysis by Product Line:')
display(price_volume_analysis)  
```

### Insights:
### Price and volume range are very simillar. There is no dominant product line or "Premium Line" by price.
### We can notice here for example that Electronic accesories generates sales by high volume but that doesn't necesairly means more money income, so that makes it a High volume line. Fashion accesories on the other hand, depends more on a higher price meaning it's products have more value.


```python
# Now we want to see what is impulsing the sales, so we will calculate the average price per unit for each product line.
# We will also calculate the quantity sold per transaction for each product line and net sales for each product line.
# We will look into the buyers habits, how much are they spending per ticket and the quantity of products they are buying per ticket. 
# Also we will look into how many product lines are being bought per ticket and how many products are being bought per ticket.
sales_per_ticket = data.groupby('invoice_id')['quantity'].sum()
sales_per_ticket = sales_per_ticket.reset_index()

avg_ticket_sales = data.groupby('invoice_id')['net_sales'].sum().mean()
avg_quantity_per_ticket = sales_per_ticket['quantity'].mean()
productlines_per_ticket = data.groupby('invoice_id')['product_line'].nunique().mean()

print(f'The average sales per ticket is: {avg_ticket_sales:.2f}')
print(f'The average items bought per ticket is: {avg_quantity_per_ticket:.2f}')
print(f'The Average number of product lines bought together per ticket is: {productlines_per_ticket}')
```

## 3. When are we selling ?

* ### At what time does the business sell more ?
* ### Which day does the business sell more?

### We will look into time patterns and behavior to see if there is a ceirtain time or day that sales are mor frequent.

### This analysis is usefull to elaborate employee shifts, marketing promotions and manage stock.

```python
# First of all we need to convert the date and time columns to datetime format.
data['date'] = pd.to_datetime(data['date'], dayfirst=False)
data['time'] = pd.to_datetime(data['time'])

# Now we create a new column for hours of the day.
data['hour'] = data['time'].dt.hour
data.head()
```

```python
# No we will lok into sales per hour. 
sales_by_hour = data.groupby('hour')['net_sales'].sum().reset_index()
sales_by_hour.sort_values('net_sales', ascending=False)
display(sales_by_hour)

plt.figure(figsize=(10,6))
sns.lineplot(data=sales_by_hour, x='hour', y='net_sales', marker='o')
plt.title('Net Sales by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Net Sales')
plt.show()
```

```python
# Amount of tickets by hour
tickets_per_hour = data.groupby('hour')['invoice_id'].nunique().reset_index()
tickets_per_hour.sort_values('invoice_id', ascending=False)
tickets_per_hour.rename(columns={'invoice_id':'tickets'}, inplace=True)
display(tickets_per_hour)
```

### Insights:
### The peak at 7:00pm indicates the moment of most sales, also more tickets, probbably afterwork purchases; also, the peak at 1:00Pm might mean a "lunchbreak"purchase.
### This is usefull for staff and stock considerations, and also for on-site promotions.

```python
# Sales per day of the week.

# First we will create a Day of purchase column. 
data['day_of_week'] = data['date'].dt.day_name() 

# Calculate sales by day of the week
sales_by_dow = data.groupby('day_of_week')['net_sales'].sum().reset_index()

# Ordering by day of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales_by_dow['day_of_week'] = pd.Categorical(sales_by_dow['day_of_week'], categories=days_order, ordered=True)
sales_by_dow = sales_by_dow.sort_values('day_of_week')

display(sales_by_dow)
```

```python
# Now we visualize the sales by day of the week.

plt.figure(figsize=(8,6))
sns.barplot(data=sales_by_dow, x='day_of_week', y='net_sales', palette='viridis')
plt.title('Net Sales by Day of The Week')
plt.xlabel('Day of The Week')
plt.ylabel('Net Sales')
plt.show()

```

### Insights:
### Here we can notice that saturday and tuesday are the days witn maximum sales.
### The business should prioritize restock before those 2 days and have enough manpower to function propperly especially at peak hours.
### Sales tough are relatively balanced across the week.
### Interestingly Tuesday also shows higher sales when compared to other days of the week; this could indicate mabe specific customer patterns or mabe there is a promotion going on those days.

### Conclusions:
### Sales show a clear weekly pattern, with Saturday generating the highest revenue, approximately 29% higher than midweek days and nearly 47% higher than Monday. This suggests stronger weekend shopping behavior, which may be associated with customer availability or typical grocery shopping habits. While Monday shows the lowest sales, potential strategies to increase traffic should consider whether promotional efforts would create new demand or simply shift purchases from other days.

## 4. Where are We Selling?

```python
# How many stores are in the dataset?
print('Number of unique stores: ', data['branch'].nunique())
```

```python
# Lets calculate the total sales per branch. 
sales_per_branch = data.groupby('branch')['net_sales'].sum().reset_index()
sales_per_branch = sales_per_branch.sort_values('net_sales', ascending=False)
print('Total Sales Per Branch:')
display(sales_per_branch)

# We vizualize the sales per branch.
plt.figure(figsize=(10,6))
sns.barplot(data=sales_per_branch, x='branch', y='net_sales', hue='branch', palette='viridis')
plt.title('Total Net Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Net Sales')
plt.show()

```

### Insights:
### This results show us that sales are highly balanced across 3 branches. There is only 4% difference between highest and lowest in sales. This could mean that they are in simillar demand zones and consistent operational performance. 
### Branches generate similar total revenue but differ in purchasing behavior, with some branches showing higher ticket values and others higher transaction counts.

### Now let's look into the average ticket per branch.

```python
# We are now going to calculate how many sales per branch.
total_transactions_per_branch = data.groupby('branch')['invoice_id'].nunique().reset_index()
print("Total Transactions Per Store:")
display(total_transactions_per_branch)
```

```python
# We calculate the average sales per ticket by Branch
# But first we need to calculate the total sales per invoice
sales_per_invoice = data.groupby(['branch', 'invoice_id'])['sales'].sum().reset_index() 
avg_ticket_per_branch = sales_per_invoice.groupby('branch')['sales'].mean().reset_index()
avg_ticket_per_branch.rename(columns={'sales': 'avg_ticket_sales'}, inplace=True)
display(avg_ticket_per_branch)

```

### Insights:
### Giza generates the highest total sales even tough having the lowest number of transactions. This is by a significantly higher average ticket size, compared to the other 2 branches. Customers in Giza spend more per purchase.

### But... Why? Let's See.

```python
# How many items per ticket ?
items_per_ticket_per_branch = data.groupby(['branch', 'invoice_id'])['quantity'].sum().reset_index()
avg_items_per_ticket_per_branch = items_per_ticket_per_branch.groupby('branch')['quantity'].mean().reset_index()
avg_items_per_ticket_per_branch.rename(columns={'quantity':'avg_items_per_ticket'},inplace=True)
avg_items_per_ticket_per_branch.sort_values('avg_items_per_ticket', ascending=False)
print('Average Items Per Ticket by Branch:')
display(avg_items_per_ticket_per_branch)
```

### Insights:
### Giza shows a slightly higher number of items per ticket, a 2% increase compared to Alex. This small difference might sugest that the higher average ticket size is not primarily driven by quantity, but mabe due to the higher priced items being purchased.

### Let's see if this is true looking into prices.

```python
# Products price analysis.
avg_product_price = data.groupby('branch')['unit_price'].mean().reset_index()
print('Average Product Price by Branch:')
display(avg_product_price)
```

### Conclusions:
### Giza generates the highest revenue despite having the fewest transactions. This is explained by a higher average ticket size, driven by both a slightly higher number of items per purchase and a higher average unit price compared to the other branches.

## 5. Sales by payment method analysis.

## Which payment method do clients chose?

```python
# We group the branches and payment method and see net sales.
sales_by_payment = data.groupby(['branch', 'payment'])['net_sales'].sum().reset_index()
sales_by_payment_pivot = sales_by_payment.pivot(index='branch', columns='payment', values='net_sales')
display(sales_by_payment_pivot)
```

```python
# Now we calculate total sales per branch from the pivot table. 
total_sales_per_branch = sales_by_payment_pivot.sum(axis=1) 
print('Total Sales Per Branch:')
display(total_sales_per_branch)
```

```python
# Now we calculate the percentage of sales by payment method for each branch. 
payment_pct = sales_by_payment_pivot.div(total_sales_per_branch, axis=0)
payment_pct = (payment_pct * 100).reset_index()
print('Percentage of Sales by Payment Method for Each Branch:')
display(payment_pct)
```

### Insights:
### Here we can see that payment preferences vary across branches. This suggests different customer payment behaviors depending on the branch location.


```python
# Payment method visualization. 
payment_pct.plot(kind='bar', stacked=True, figsize=(8,5), title='Payment Method Distribution by Branch')
plt.ylabel('Percentage of Sales')
plt.show()
```

#### Payment method Insights:
#### Payment preferences vary by branch, suggesting different customer profiles across locations.
